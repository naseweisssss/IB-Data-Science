## 2b) Fit a linear model $f(x) = \beta_0 + \beta_1 x$ to the data

- Print the values of the regression coefficients $\beta_0, \beta_1$ determined using least-squares. 
- Plot the fitted model and the scatterplot on the same plot.
- Compute and print the **MSE** and the $R^2$ coefficient for the fitted model.

All numerical outputs should be displayed to three decimal places.

```python
import numpy as np
import matplotlib.pyplot as plt

# load data
yvals = np.array(df_tfl['Tube Total (000s)'])
N = np.size(yvals)
xvals = np.linspace(1,N,N)

# fit a linear model
beta = np.polyfit(xvals, yvals, deg=1)
beta_0, beta_1 = beta[1], beta[0]
print(f"Regression coefficients: beta_0 = {beta_0:.3f}, beta_1 = {beta_1:.3f}")

# compute the predicted values
y_pred = beta_0 + beta_1 * xvals

# plot the fitted model and the scatterplot
plt.scatter(xvals, yvals, label='data')
plt.plot(xvals, y_pred, color='red', label='linear model')
plt.title('Tube Total (000s) from 1 Apr 2000 to 2 Feb 2019')
plt.xlabel('Period')
plt.ylabel('Tube Total (000s)')
plt.legend()
plt.show()

# compute and print the MSE and R^2
MSE = np.mean((y_pred - yvals) ** 2)
R2 = 1 - MSE / np.var(yvals)
print(f"MSE: {MSE:.3f}")
print(f"R^2 coefficient: {R2:.3f}")
```

## 2d) Periodogram

- Compute and plot the peridogram of the residuals. (Recall that the periodogram is the squared-magnitude of the DFT coefficients.)
- Identify the indices/frequencies for which the periogram value exceeds **50%** of the maximum. 


```python
import numpy as np
from scipy.signal import periodogram
import matplotlib.pyplot as plt

# load data and fit linear model
yvals = np.array(df_tfl['Tube Total (000s)'])
N = np.size(yvals)
xvals = np.linspace(1,N,N)
beta = np.polyfit(xvals, yvals, deg=1)
beta_0, beta_1 = beta[1], beta[0]

# compute the predicted values and residuals
y_pred = beta_0 + beta_1 * xvals
residuals = yvals - y_pred

# compute the periodogram
frequencies, periodogram_values = periodogram(residuals)

# plot the periodogram
plt.plot(frequencies, periodogram_values)
plt.title('Periodogram of Residuals')
plt.xlabel('Frequency')
plt.ylabel('Periodogram')
plt.show()


```

```python
# find indices/frequencies for which the periodogram exceeds 50% of the maximum
max_periodogram = np.max(periodogram_values)
threshold = 0.5 * max_periodogram
indices_above_threshold = np.where(periodogram_values > threshold)[0]
frequencies_above_threshold = frequencies[indices_above_threshold]

print('Frequencies for which the periodogram exceeds 50% of the maximum:')
print(frequencies_above_threshold)
```

## 2e) To the residuals,  fit a model of the form  

$$ \beta_{1s} \sin(\omega_1 x) + \beta_{1c} \cos(\omega_1 x) + \beta_{2s} \sin(\omega_2 x) + \beta_{2c} \cos(\omega_2 x) + \ldots + \beta_{Ks} \sin(\omega_K x) + \beta_{Kc} \cos(\omega_K x).$$

The frequencies $\omega_1, \ldots, \omega_K$ in the model are those corresponding to the indices identified in Part 2c. (Hint: Each of the sines and cosines will correspond to one column in your X-matrix.)

- Print the values of the regression coefficients obtained using least-squares. 
- Compute and print the final **MSE** and $R^2$ coefficient. Comment on the improvement over the linear fit.

All numerical outputs should be displayed to three decimal places.

```python
from scipy.signal import find_peaks

# Fit linear model
yvals = np.array(df_tfl['Tube Total (000s)'])
N = np.size(yvals)
xvals = np.linspace(1,N,N)
y_bar = np.mean(yvals)
X = np.vstack((np.ones(N), xvals)).T
beta, _, _, _ = np.linalg.lstsq(X, yvals, rcond=None)

# Compute predictions and residuals
y_pred = X @ beta
residuals = yvals - y_pred

# Plot scatterplot and linear fit
plt.scatter(xvals, yvals, s=5, label='Data')
plt.plot(xvals, y_pred, 'r', label='Linear fit')
plt.legend()
plt.show()

# Compute MSE and R-squared for linear fit
MSE = np.mean(residuals**2)
R2 = 1 - np.sum(residuals**2)/np.sum((yvals - y_bar)**2)
print('Linear Model')
print(f'beta_0 = {beta[0]:.3f}')
print(f'beta_1 = {beta[1]:.3f}')
print(f'MSE = {MSE:.3f}')
print(f'R^2 = {R2:.3f}')

# Plot residuals
plt.scatter(xvals, residuals, s=5)
plt.xlabel('Time (days)')
plt.ylabel('Residuals')
plt.show()
```

## 2f) The combined fit
- Plot the combined fit together with a scatterplot of the data
- Compute and print the final **MSE** and $R^2$ coefficient. Comment on the improvement over the linear fit.

The combined fit, which corresponds to the full model

$$
f(x) = \beta_0 + \beta_1 x + \beta_{s1} \sin(\omega_1 x) + \beta_{c1} \cos(\omega_1 x) + \ldots + \beta_{sk} \sin(\omega_k x) + \beta_{ck} \cos(\omega_k x),
$$

can be obtained by adding the fits in parts 2b) and 2e).

```python
# Your code here
# Compute combined fit
freqs = 2*np.pi*peaks/N
X = np.vstack((np.ones(N), xvals, 
               np.sin(np.outer(xvals, freqs)).T, 
               np.cos(np.outer(xvals, freqs)).T)).T
beta, _, _, _ = np.linalg.lstsq(X, yvals, rcond=None)
y_pred = X @ beta
residuals = yvals - y_pred

# Plot combined fit and data
plt.scatter(xvals, yvals, s=5, label='Data')
plt.plot(xvals, y_pred, 'r', label='Combined fit')
plt.legend()
plt.show()

# Compute MSE and R-squared for combined fit
MSE = np.mean(residuals**2)
R2 = 1 - np.sum(residuals**2)/np.sum((yvals - y_bar)**2)
print('Combined Fit')
print(f'beta_0 = {beta[0]:.3f}')
print(f'beta_1 = {beta[1]:.3f}')
for i in range(len(freqs)):
    print(f'beta_s{i+1} = {beta[2*i+2]:.3f}')
    print(f'beta_c{i+1} = {beta[2*i+3]:.3f}')
print(f'MSE = {MSE:.3f}')
print(f'R^2 = {R2:.3f}')
```
