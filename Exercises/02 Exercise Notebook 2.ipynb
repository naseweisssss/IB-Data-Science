## 2b) Fit an exponential model to the data 

From our knowledge of how the virus spreads, we know that the number of infections, hospital admissions etc. should (roughly) follow an exponential curve. We would therefore like to fit a model of the form $y = c_1 e^{c_2 x}$, where $y$ is the number of admissions on day $x$. 

Note that this is a linear model on a log-scale for $y$. That is, $\log y = \log c_1 +  c_2 x$.

- Fit a linear model for $\log(\text{yvals})$ vs $\text{xvals}$, and print the values of $c_1$ and $c_2$
- Plot the fit $y = c_1 e^{c_2x}$ along with the scatterplot of the data

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

# Transform data by taking natural logarithm
yvals_ln = np.log(yvals)

# Construct design matrix
X = np.ones((N, 2))
for i in range(N):
    X[i, 0] = 1
    X[i, 1] = xvals[i]

# Fit linear model using least squares
beta = np.linalg.lstsq(X, yvals_ln, rcond=None)[0]

# Extract c1 and c2 from the estimated coefficients
c1 = np.exp(beta[0])
c2 = beta[1]

# Print values of c1 and c2
print('c1 =', c1)
print('c2 =', c2)

# Create array of x-values for the fitted curve
x_fit = np.linspace(1, N, 1000)

# Evaluate the fitted curve at the x-values
y_fit = c1 * np.exp(c2 * x_fit)

# Plot scatter plot of the data
plt.scatter(xvals, yvals, label='data')

# Plot the fitted curve
plt.plot(x_fit, y_fit, label='fit')

# Add labels and legend
plt.xlabel('xvals')
plt.ylabel('yvals')
plt.title('Exponential fit of yvals')
plt.legend()

# Show the plot
plt.show()
```

## 2c) Estimate the weekly growth rate in hospital admissions (in \%) over this period

_Hint_ : According to the model, admissions increase every 7 days by a factor of 
$(c_1 e^{c_2 (x+7)})/(c_1 e^{c_2 x}) = e^{7c_2}$.

```python
# compute and print weekly growth rate (in %)
# Compute factor by which hospital admissions increase every 7 days
factor = np.exp(7 * c2)

# Compute weekly growth rate in percent
growth_rate = 100 * (factor - 1)

# Print estimated weekly growth rate
print('Estimated weekly growth rate in hospital admissions: {:.2f}%'.format(growth_rate))
```
